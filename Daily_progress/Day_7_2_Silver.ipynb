```python
from pyspark.sql import functions as F
```

```python
month = "Oct"
```

```python
df_bronze = spark.read.table("workspace.bronze.ecommerce_raw")
```

```python
df_silver = df_bronze.filter("user_id IS NOT NULL") \
                     .dropDuplicates(["event_time", "user_id", "product_id"]) \
                     .withColumn("brand", F.lower(F.col("brand")))

df_silver.write.format("delta").mode("overwrite").saveAsTable("ecommerce_silver")
```
