```python
from pyspark.sql import functions as F
```

```python
month = "Oct"
```

```python
df_silver = spark.read.table("ecommerce_silver")
```

```python
df_gold = df_silver.filter("event_type = 'purchase'") \
                   .groupBy("brand", F.window("event_time", "1 day")) \
                   .agg(F.sum("price").alias("daily_revenue"))
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("ecommerce_gold")
```
